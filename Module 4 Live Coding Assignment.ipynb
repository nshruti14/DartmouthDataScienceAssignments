Question 1
Define a function called ‚Äúcalculate_si_amount‚Äù that takes the parameters ‚Äúprincipal‚Äù, ‚Äúrate‚Äù, and ‚Äútime‚Äù and computes the ‚Äúinterest‚Äù using the formula interest = principal*rate*time/100,  and returns the variable ‚Äútotal‚Äù computed by adding ‚Äúprincipal‚Äù and ‚Äúrate‚Äù. Next, call the function using the values principal = 1859, rate = 1.67, and time = 5.

```python
def calculate_si_amount(principal, rate, time):
    interest=(principal*rate*time)/100
    total=principal+interest
    return total
calculate_si_amount(1859, 1.67, 5)
```

Import the public ‚Äúdiabetes‚Äù database from the sklearn library and save it to a dataframe called ‚Äúdiabetes‚Äù. Print to screen the shape of the dataset and the first ten lines of the dataset.
```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

diabetes_dataset=load_diabetes()
diabetes=pd.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
diabetes['target']=diabetes_dataset.target
display(diabetes.head(10), diabetes.shape)
```

```python
X=diabetes[['age']]
y=diabetes['bmi']
regr=LinearRegression()
regr.fit(X,y)
print('Intercept'+ str(regr.intercept_))
print('Coefficient' +str( regr.coef_))
print(regr.score(X,y))
```

Question 4
Import the public ‚Äúdiabetes‚Äù database from the sklearn library and save it to a dataframe called ‚Äúdiabetes‚Äù. Next, perform a multiple linear regression using the variables ‚ÄúAge‚Äù, ‚ÄúBody mass index‚Äù to predict the variable ‚ÄúAverage blood pressure‚Äù. Save your model in a variable called ‚Äúmultiple prediction‚Äù. Print to screen the intercept and the coefficients obtained from your linear regression model and the score of your prediction.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

diabetes_dataset=load_diabetes()
diabetes=pd.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
diabetes['target']=diabetes_dataset.target

X=diabetes[['age', 'bmi']]
y=diabetes['bp']

multiple_prediction=LinearRegression()
multiple_prediction.fit(X,y)

print('Intercept '+ str(multiple_prediction.intercept_))
print('Coefficient ' +str(multiple_prediction.coef_))
print('Score '+ str(multiple_prediction.score(X,y)))
```

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np

housing={'Year':[2017,2017,2017,2017,2018,2018,2018,2019,2019],'Income(K)':[67,78,69,130,95,113,99,178,156],'Bedrooms':[2,1,1,2,2,2,3,4,3],'Bathrooms':[1,1,2,2,1,2,1,3,2]}
df=pd.DataFrame(housing)

df.head(5)
```

BONUS: Import the Boston dataset from sklearn "load_boston", and predict the variable "MEDV" using the columns "LSTAT", "INDUS", "TAX", "RM", "CRIM". Save your model into a variable called boston_model, fit your model and print to the screen your intercept, coeficcients and the Mean Absolute Error (MAE).

Hint: If you use score you'll be getting the  ùëÖ2 , you need to look for the MAE, or define it‚Ä©

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

boston_dataset=load_boston()
boston=pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV']=boston_dataset.target
display(boston)
```

```python
X=boston[['LSTAT', 'INDUS', 'TAX', 'RM', 'CRIM']]
y=boston['MEDV']

regr=LinearRegression()
regr.fit(X,y)
y_pred=regr.predict(X)

print('Intercept'+ str(regr.intercept_))
print('Coefficient' +str( regr.coef_))
print(mean_absolute_error(y,y_pred))
```
